# Approximate the Night Time of a Given Latitude

This approach takes into consideration one longitude sliver and multiple latitudes of high resoution 

One Factor to consider was dealing with optical observations is the approximate amount of night time that there is a given month. 

```python
import numpy as np
import pytz
import datetime
from astropy import units as u
from astropy.coordinates import EarthLocation
from astroplan import Observer
from astropy.time import Time
import matplotlib.pyplot as plt
from datetime import timedelta
from timezonefinder import TimezoneFinder
```

```python
location = EarthLocation.from_geodetic(lon=0*u.deg, lat=0*u.deg, height=0*u.m)
observer = Observer(location=location)
```

```python
# Define the longitude and Latitude grids
lons = np.arange(-180, 180,20 )
lats = np.arange(-90, 90, 10)

# Calculate the average night duration for each location for each day of the month
avg_night_duration = np.zeros((len(lats), len(lons), 30))
```

```python
year=2022
month=1
day=15
tf=TimezoneFinder()
night_duration_sum=np.zeros([len(lats),len(lons)])
original_dt = datetime.datetime(year, month, day, 0, 0, 0, tzinfo=pytz.utc)
sunrise_times = np.zeros((len(lats), len(lons)))
sunset_times = np.zeros((len(lats), len(lons)))
sunlit=  np.zeros((len(lats), len(lons)))
for i, lat in enumerate(lats):
    for j, lon in enumerate(lons):
        # Find the new timezeon based on lat and long
        # Find the timezone based on the latitude and longitude
        timezone_str = tf.timezone_at(lat=lat, lng=lon)
        target_timezone = pytz.timezone(timezone_str)
        dt=original_dt.astimezone(target_timezone)
        
        
        # Define the observer location
        location = EarthLocation.from_geodetic(lon=lon*u.deg, lat=lat*u.deg, height=0*u.m)
        observer = Observer(location=location)

        # Calculate the sunrise and sunset times
        #TODO: Add components of astronomical twilight


        # Pole Conditions (where sun may not rise/set)
        if observer.is_night(Time(dt)):
            
        
            try:
                sunrise_time = observer.sun_rise_time(Time(dt), which='next').datetime
                sunset_time = observer.sun_set_time(Time(dt), which='previous').datetime


            except:
                if observer.is_night(Time(dt)) ==True:
                    sunrise_time = dt+timedelta(days=1)
                    sunset_time = dt
                elif observer.is_night(Time(dt)) == False: # Is day
                    sunrise_time = dt
                    sunset_time = dt
        else:
            try:
                sunrise_time = observer.sun_rise_time(Time(dt), which='previous').datetime
                sunset_time = observer.sun_set_time(Time(dt), which='previous').datetime


            except:
                if observer.is_night(Time(dt)) ==True:
                    sunrise_time = dt+timedelta(days=1)
                    sunset_time = dt
                elif observer.is_night(Time(dt)) == False: # Is day
                    sunrise_time = dt
                    sunset_time = dt
            
            


        sunrise_times[i, j] = sunrise_time.timestamp()
        sunset_times[i, j] = sunset_time.timestamp()
        sunlit[i,j]=observer.is_night(Time(dt))
night_duration=sunrise_times-sunset_times
#night_duration=abs(night_duration)
night_duration[night_duration<0]=0
night_duration_sum += night_duration

```

```python
def calculate_average_night_time(month, day, year, lats, lons):
    original_dt = datetime.datetime(year, month, day, 0, 0, 0, tzinfo=pytz.utc)
    night_duration_sum = np.zeros([len(lats), len(lons)])
    sunrise_times = np.zeros((len(lats), len(lons)))
    sunset_times = np.zeros((len(lats), len(lons)))
    
    for i, lat in enumerate(lats):
        for j, lon in enumerate(lons):
            # Find the new timezone based on lat and long
            # Find the timezone based on the latitude and longitude
            timezone_str = tf.timezone_at(lat=lat, lng=lon)
            target_timezone = pytz.timezone(timezone_str)
            dt = original_dt.astimezone(target_timezone)
            #dt=original_dt
            # Define the observer location
            location = EarthLocation.from_geodetic(lon=lon*u.deg, lat=lat*u.deg, height=0*u.m)
            observer = Observer(location=location)

            # Calculate the sunrise and sunset times
            # TODO: Add components of astronomical twilight
            
            # Pole Conditions (where sun may not rise/set)
            if observer.is_night(Time(dt)):
            
        
                try:
                    sunrise_time = observer.sun_rise_time(Time(dt), which='next').datetime
                    sunset_time = observer.sun_set_time(Time(dt), which='previous').datetime


                except:
                    if observer.is_night(Time(dt)) ==True:
                        sunrise_time = dt+timedelta(days=1)
                        sunset_time = dt
                    elif observer.is_night(Time(dt)) == False: # Is day
                        sunrise_time = dt
                        sunset_time = dt
            else:
                try:
                    sunrise_time = observer.sun_rise_time(Time(dt), which='previous').datetime
                    sunset_time = observer.sun_set_time(Time(dt), which='previous').datetime


                except:
                    if observer.is_night(Time(dt)) ==True:
                        sunrise_time = dt+timedelta(days=1)
                        sunset_time = dt
                    elif observer.is_night(Time(dt)) == False: # Is day
                        sunrise_time = dt
                        sunset_time = dt





            sunrise_times[i, j] = sunrise_time.timestamp()
            sunset_times[i, j] = sunset_time.timestamp()
    night_duration=sunrise_times-sunset_times
    night_duration[night_duration < 0] = 0
    night_duration_sum += night_duration
    
    return night_duration_sum / (60*60*24)
```

```python
import rasterio
def save_tiff(array,dst_file):
    dst_crs = "EPSG:4326"  # coordinate reference system
    dst_transform = rasterio.transform.from_bounds(-180, -90, 180, 90, 3600, 1800)
    with rasterio.open(
        dst_file,'w+',driver='GTiff',
        height=array.shape[0],
        width=array.shape[1],
        count=1, 
        dtype=array.dtype, 
        crs=dst_crs, 
        transform=dst_transform) as dst:
        dst.write(array,1)
```

```python
def calculate_average_night_time_with_elevation(month, day, year, lats, lons,DEM_array,step):
    original_dt = datetime.datetime(year, month, day, 0, 0, 0, tzinfo=pytz.utc)
    night_duration_sum = np.zeros([len(lats), len(lons)])
    sunrise_times = np.zeros((len(lats), len(lons)))
    sunset_times = np.zeros((len(lats), len(lons)))
    lat_scale=np.shape(DEM_array)[0]/180
    lon_scale=np.shape(DEM_array)[1]/360
    
    for i, lat in enumerate(lats):
        for j, lon in enumerate(lons):
            # Find the new timezone based on lat and long
            # Find the timezone based on the latitude and longitude
            timezone_str = tf.timezone_at(lat=lat, lng=lon)
            target_timezone = pytz.timezone(timezone_str)
            dt = original_dt.astimezone(target_timezone)

            # Define the observer location
            
            if DEM_array.mask[int((lat+90)*lat_scale),int((lon+180)*lon_scale)]== True:
                el=0
            else:
            
                el=DEM_array[int((lat+90)*lat_scale),int((lon+180)*lon_scale)]
            
            
            observer = Observer(longitude=lon*u.deg, 
                                latitude=lat*u.deg,
                                elevation=el*u.m)
            

            if observer.is_night(Time(dt)):
            
        
                try:
                    sunrise_time = observer.sun_rise_time(Time(dt), which='next').datetime
                    sunset_time = observer.sun_set_time(Time(dt), which='previous').datetime


                except:
                    if observer.is_night(Time(dt)) ==True:
                        sunrise_time = dt+timedelta(days=1)
                        sunset_time = dt
                    elif observer.is_night(Time(dt)) == False: # Is day
                        sunrise_time = dt
                        sunset_time = dt
            else:
                try:
                    sunrise_time = observer.sun_rise_time(Time(dt), which='next').datetime
                    sunset_time = observer.sun_set_time(Time(dt), which='next').datetime


                except:
                    if observer.is_night(Time(dt)) ==True:
                        sunrise_time = dt+timedelta(days=1)
                        sunset_time = dt
                    elif observer.is_night(Time(dt)) == False: # Is day
                        sunrise_time = dt
                        sunset_time = dt
        
            sunrise_times[i, j] = sunrise_time.timestamp()
            sunset_times[i, j] = sunset_time.timestamp()

    night_duration = sunrise_times - sunset_times
    night_duration[night_duration < 0] = 0
    night_duration_sum += night_duration
    
    return night_duration_sum / (60*60*24)
```

There are some outliers in the data. A solution to this is to find outliers and replace them with a rolling average.

```python
threshold=1
for i in range(0,np.shape(Normalized_Time_Night_03_with_elevation)[0]):
    mean=np.mean(Normalized_Time_Night_03_with_elevation[i,:])
    std=np.std(Normalized_Time_Night_03_with_elevation[i,:])
    
    lower_bound=mean-threshold*std
    upper_bound=mean+threshold*std
    
    for j in range(0,np.shape(Normalized_Time_Night_03_with_elevation)[1]):
        if Normalized_Time_Night_03_with_elevation[i,j] < lower_bound or Normalized_Time_Night_03_with_elevation[i,j] > upper_bound:
            
            moving_mean= np.mean([Normalized_Time_Night_03_with_elevation[i,j-1],Normalized_Time_Night_03_with_elevation[i,j+1]])
            Normalized_Time_Night_03_with_elevation[i,j]=moving_mean
    
```

```python
def apply_rolling_average (array):
    threshold=0.5
    copy=array.copy()
    
    for i in range(0,np.shape(copy)[0]):
            mean=np.mean(copy[i,:])
            std=np.std(copy[i,:])



            lower_bound=mean-threshold*std
            upper_bound=mean+threshold*std

            line=(copy[i,:]).copy()
            valid_x_coords=np.where((line > lower_bound) & (line < upper_bound))
            line_y=line[valid_x_coords]
            interp_x=np.where((line < lower_bound) | (line > upper_bound))

            if len(interp_x[0])>0:
                y_interp=np.interp(interp_x[0],valid_x_coords[0],line_y)
                for l,x in enumerate(interp_x[0]):

                    line[x]=y_interp[l]
                copy[i,:]=line
    return copy
    
```

```python
import rasterio
def save_tiff(array,dst_file):
    dst_crs = "EPSG:4326"  # coordinate reference system
    dst_transform = rasterio.transform.from_bounds(-180, -90, 180, 90, np.shape(array)[0], np.shape(array)[1])
    with rasterio.open(
        dst_file,'w+',driver='GTiff',
        height=array.shape[0],
        width=array.shape[1],
        count=1, 
        dtype=array.dtype, 
        crs=dst_crs, 
        transform=dst_transform) as dst:
        dst.write(array,1)
```
